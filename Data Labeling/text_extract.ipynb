```python
import json
import os

def parse_interview_simple(file_path, interviewee, interviewers):
    """
    Parses interview transcripts into Q&A pairs for scoring

    :param file_path: Path to the transcript .txt file.
    :param interviewee: The interviewee's name.
    :param interviewers: A list of interviewer names.
    :return: A list of Q&A pairs with names included.
    """

    with open(file_path, "r", encoding="utf-8") as f:
        lines = [line.strip() for line in f if line.strip()] 

    qa_pairs = []
    i = 0

    while i < len(lines) - 2:
        if lines[i] in interviewers and lines[i + 2] == interviewee:
            qa_pairs.append({
                "interviewer": lines[i],
                "question": lines[i + 1],
                "interviewee": lines[i + 2],
                "response": lines[i + 3] if i + 3 < len(lines) else "",
                "follow-up": lines[i + 5] if i + 5 < len(lines) else ""
            })
            i += 4  # Move to next Q&A block
        else:
            i += 1  # Keep scanning if pattern isn't found

    return qa_pairs
```

```python
interviewers=['Juan Manuel Benítez', 'Nicole Gelinas', 'Josh Greenman', 'Alyssa Katz', 'Ben Max', 'Myles Miller', 'Harry Siegel', 'Ben Smith', 'Liena Zagare']
out1=parse_interview_simple('test1.md','Brad Lander', interviewers)
```
